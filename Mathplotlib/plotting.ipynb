```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y,color="red")
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
c=np.array(["green","black","hotpink","blue"])
plt.barh(x, y,color=c,height=0.4)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
myexplode = np.array([0.2, 0, 0, 0])
plt.pie(y, labels = mylabels, colors = mycolors, explode = myexplode, shadow = True,startangle=180)
plt.legend()
plt.show() 
```
