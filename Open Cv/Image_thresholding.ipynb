## **Thresholding**
Thresholding is an image segmentation method. It compares pixel values with a threshold value and updates it accordingly. OpenCV supports multiple variations of thresholding. 

Thresholding can only be applied to grayscale images.It is Binarising images

```python
import cv2 as cv
```

```python
img=cv.imread("pic1.webp")
img_gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

### **Simple Image Thresholding**

```python
threshold,thresh_binary=cv.threshold(img_gray,127,255,cv.THRESH_BINARY) # If pixel is > 127 then it wil be 255
threshold,thresh_binary_inv =cv.threshold(img_gray,127,255,cv.THRESH_BINARY_INV)
threshold,thresh_trunc = cv.threshold(img_gray,127,255,cv.THRESH_TRUNC)
threshold,thresh_tozero =cv.threshold(img_gray,127,255,cv.THRESH_TOZERO)
threshold,thresh_tozero_inv = cv.threshold(img_gray,127,255,cv.THRESH_TOZERO_INV)
```

```python
names = ['Original Image','BINARY','THRESH_BINARY_INV','THRESH_TRUNC','THRESH_TOZERO','THRESH_TOZERO_INV']
images = [img_gray,thresh_binary,thresh_binary_inv,thresh_trunc,thresh_tozero,thresh_tozero_inv]
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i])
    plt.title(names[i])
    plt.xticks([]),plt.yticks([])
    
#plt.show()
    
    
```
